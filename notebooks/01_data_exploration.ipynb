Loading Data

```python
# üì¶ Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ü™ê Load Data
df = pd.read_csv("../data/kepler_data.csv")
df.head()
```

```python
# üëÄ Basic Overview
print("Shape:", df.shape)
print("\nColumns:\n", df.columns.tolist())
df.info()
```

Checking for missing values

```python
# üîç Missing Values
plt.figure(figsize=(10,5))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
print("Missing Values per COlumn:\n")
df.isnull().sum().sort_values(ascending=False).head(10)
```

```python
# üéØ Target Distribution
plt.figure(figsize=(6,4))
sns.countplot(x='exoplanet_archive_disposition', data=df)
plt.title("Target Class Distribution")
plt.show()
```

```python
# ‚öôÔ∏è Correlation Matrix
numeric_df = df.select_dtypes(include=['float64', 'int64'])
corr = numeric_df.corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# üå°Ô∏è Example: Relationship between Planet Radius and Equilibrium Temperature
sns.scatterplot(x='koi_prad', y='koi_eqtemp', 
                hue='exoplanet_archive_disposition', data=df)
plt.title("Planet Size vs. Temperature")
plt.show()
```

```python
# Drop non-predictive / target-leaking columns ===
drop_cols = [
    'kepid',
    'kepler_name ',
    'disposition_score',
    'koi_tce_plnt_num',
    'koi_tce_delivname',
]
df.drop(columns=[c for c in drop_cols if c in df.columns], inplace=True)

print("Data shape after dropping columns:", df.shape)
df.head()
```

```python
from sklearn.impute import SimpleImputer
# Impute numeric columns using median
numeric_cols = [
    'koi_eqtemp',
    'koi_srad',
    'koi_slogg',
    'koi_stefftemp',
    'koi_snr',
    'koi_param',
    'koi_depth',
    'koi_prad']
numeric_imputer = SimpleImputer(strategy='median')
df[numeric_cols] = numeric_imputer.fit_transform(df[numeric_cols])

# Save the imputed dataset
df.to_csv("../data/kepler_data_cleaned.csv", index=False)
print(f"‚úÖ Dataset saved with imputed missing values. Shape: {df.shape}")
```
